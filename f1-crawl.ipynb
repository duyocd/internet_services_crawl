```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import csv
from datetime import datetime
```

```python
url = 'https://www.formula1.com/en/results/2024/races/1252/abu-dhabi/race-result'
response = requests.get(url)
response
```

```python
csv_file = '2024_season_result.csv'
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table', class_='f1-table f1-table-with-data w-full')
    rows = table.find_all('tr')
    
    with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Position','Driver','Car','Laps','Points'])
        
        for row in rows[:-1]:
            column = row.find_all('td')
            if column:
                position = column[0].text.strip()
                driver = column[2].text.strip()
                car = column[3].text.strip()
                laps = column[4].text.strip()
                point = column[6].text.strip()
                
                writer.writerow([position, driver, car, int(laps), int(point)])
```

```python
def crawl_race_result(year_start, year_end, csv_file):
    with open(csv_file, 'w', newline='', encoding='utf-8') as file: 
        writer = csv.writer(file)
        writer.writerow(['Year','Position','Driver','Car','Laps','Points'])
        for year in range(year_start, year_end):
            url = f'https://www.formula1.com/en/results/{year}/drivers'
            response = requests.get(url)
            if response.status_code == 200:
                soup = BeautifulSoup(response.content, 'html.parser')
                table = soup.find('table', class_ = 'f1-table f1-table-with-data w-full')
                rows = table.find_all('tr')
                for row in rows:
                    column = row.find_all('td')
                    if column:
                        position = column[0].text.strip()
                        driver = column[1].text.strip()
                        nationality = column[2].text.strip()
                        car = column[3].text.strip()
                        point = column[4].text.strip()
                                    
                        writer.writerow([year, position, driver, nationality, car, float(point)])
                print(f'=====INSERT DATA SUCESSFULL FOR YEAR {year}=====')
            else:
                print("ERROR! CAN NOT GET RESPONSE FROM THIS HTML")
```

```python
def crawl_race_result_by_year(year_start, year_end, csv_file):
    with open(csv_file, 'w', newline='', encoding='utf-8') as file: 
        writer = csv.writer(file)
        writer.writerow(['Year','Grand Prix','Date','Winner','Car','Laps'])
        for year in range(year_start, year_end):
            url = f'https://www.formula1.com/en/results/{year}/races'
            response = requests.get(url)
            if response.status_code == 200:
                soup = BeautifulSoup(response.content, 'html.parser')
                table = soup.find('table', class_ = 'f1-table f1-table-with-data w-full')
                rows = table.find_all('tr')
                for row in rows:
                    column = row.find_all('td')
                    if column:
                        grand_prix = column[0].text.strip()
                        date_str = column[1].text.strip()
                        try:
                            date = datetime.strptime(date_str, '%d %B %Y').date()
                        except ValueError:
                            date = date_str
                        winner = column[2].text.strip()
                        car = column[3].text.strip()
                        laps = column[4].text.strip()
                        if laps == '':
                            continue 
                        writer.writerow([year, grand_prix, date, winner, car, int(laps)])
                print(f'=====INSERT DATA SUCESSFULL FOR YEAR {year}=====')
            else:
                print("ERROR! CAN NOT GET RESPONSE FROM THIS HTML")
```

```python
crawl_race_result_by_year(1950, 2025, '1950_to_2024_race_results.csv')
```
